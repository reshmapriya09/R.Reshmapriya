```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense,Dropout
from sklearn.preprocessing import LabelEncoder,StandardScaler
from sklearn.model_selection import train_test_split
```

```python
x = df.drop("monthly_rent_inr", axis=1)
y = df["monthly_rent_inr"]
```

```python
x = pd.get_dummies(x, drop_first=True)
```

```python
print(x.select_dtypes(include="object").columns)
```

```python
cols = ["city", "locality", "property_type", "furnishing", "maintenance_included"]

for col in cols:
    df[col] = LabelEncoder().fit_transform(df[col].astype(str))
```

```python
x_train, x_test, y_train, y_test = train_test_split(
    x, y, test_size=0.2, random_state=42
)
```

```python
# feature scaling

from sklearn.preprocessing import StandardScaler

scaler_x = StandardScaler()

x_train = scaler_x.fit_transform(x_train)
x_test = scaler_x.transform(x_test)

```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
import numpy as np

lr = LinearRegression()
lr.fit(x_train, y_train)

y_pred_lr = lr.predict(x_test)

mse_lr = mean_squared_error(y_test, y_pred_lr)
rmse_lr = np.sqrt(mse_lr)
r2_lr = r2_score(y_test, y_pred_lr)

print(f"MSE = {mse_lr:.2f}, RMSE = {rmse_lr:.2f}, R2 = {r2_lr:.3f}")
```

```python
lr
```

```python
from sklearn.ensemble import RandomForestRegressor

rf = RandomForestRegressor(random_state=42)
rf.fit(x_train, y_train)

y_pred_rf = rf.predict(x_test)

mse_rf = mean_squared_error(y_test, y_pred_rf)
rmse_rf = np.sqrt(mse_rf)
r2_rf = r2_score(y_test, y_pred_rf)

print(f"MSE = {mse_rf:.2f}, RMSE = {rmse_rf:.2f}, R2 = {r2_rf:.3f}")
```

```python
rf
```

```python
# ANN model
# scale target

scaler_y = StandardScaler()

y_train_s = scaler_y.fit_transform(y_train.values.reshape(-1,1))
y_test_s = scaler_y.transform(y_test.values.reshape(-1,1))
```

```python
# build model

model = Sequential()

model.add(Dense(128, activation='relu'))
model.add(Dense(64, activation='relu'))
model.add(Dense(32, activation='relu'))
model.add(Dense(1))

```

```python
from tensorflow.keras.optimizers import Adam

model.compile(optimizer=Adam(learning_rate=0.001), loss='mse')
```

```python
# train model

from tensorflow.keras.callbacks import EarlyStopping
early = EarlyStopping(patience=5, restore_best_weights=True)

model.fit(
    x_train, y_train_s,
    epochs=50,
    validation_data=(x_test, y_test_s),
    callbacks=[early]
)
```

```python
model = Sequential()

model.add(Dense(128, activation='relu', input_shape=(x_train.shape[1],)))
model.add(Dense(64, activation='relu'))
model.add(Dense(32, activation='relu'))
model.add(Dense(1))

model.compile(optimizer='adam', loss='mse')
```

```python
model.summary()
```

```python
# dropout
model = Sequential()

model.add(Dense(128, activation='relu', input_shape=(x_train.shape[1],)))
model.add(Dropout(0.2))   

model.add(Dense(64, activation='relu'))
model.add(Dense(32, activation='relu'))
model.add(Dense(1))
```

```python
# earlystopping

early = EarlyStopping(patience=5, restore_best_weights=True)

model.compile(optimizer='adam', loss='mse')

model.fit(
    x_train, y_train_s,
    epochs=100,
    validation_data=(x_test, y_test_s),
    callbacks=[early]
)
```

```python
# evaluate ann

y_pred_s = model.predict(x_test)
y_pred_ann = scaler_y.inverse_transform(y_pred_s)

mse_ann = mean_squared_error(y_test, y_pred_ann)
rmse_ann = np.sqrt(mse_ann)
r2_ann = r2_score(y_test, y_pred_ann)

print(f"MSE = {mse_ann:.2f}, RMSE = {rmse_ann:.2f}, R2 = {r2_ann:.3f}")
```

```python
# model comparison

comparison_df = pd.DataFrame({
    "Model": ["Linear Regression", "Random Forest", "ANN"],
    "MSE": [mse_lr, mse_rf, mse_ann],
    "RMSE": [rmse_lr, rmse_rf, rmse_ann],
    "R2 Score": [r2_lr, r2_rf, r2_ann]
})

comparison_df.round(3)
```

```python
# model comparison chart

models = ["Linear Regression", "Random Forest", "ANN"]
r2_scores = [r2_lr, r2_rf, r2_ann]

plt.figure()
plt.bar(models, r2_scores)
plt.xlabel("Models")
plt.ylabel("R2 Score")
plt.title("Model Comparison (R2 Score)")
plt.show()
```

```python
# RSME comparison

rmse_values = [rmse_lr, rmse_rf, rmse_ann]

plt.figure()
plt.bar(models, rmse_values,color='yellow')
plt.xlabel("Models")
plt.ylabel("RMSE")
plt.title("Model Comparison (RMSE)")
plt.show()
```

```python
# conclusion
# Random Forest outperformed other models due to its ability to model non-linear relationships and interactions between features. 
# Linear Regression was limited by its linear assumptions.
# while ANN required more tuning and data to achieve comparable performance.
```
